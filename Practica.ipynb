```python
import json
from collections import Counter
import matplotlib.pyplot as plt

# Leer el archivo JSON
with open('NoticiasF.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extraer la información de las fechas de los artículos
fechas = []
for ciudad, articulos in data["reportePorTitulo"].items():
    for articulo in articulos:
        fecha = articulo["fecha"]
        fechas.append(fecha)

# Contar el número de artículos por fecha
conteo_por_fecha = Counter(fechas)

# Ordenar las fechas
fechas_ordenadas = sorted(conteo_por_fecha.items())

# Separar las fechas y los conteos
fechas, conteos = zip(*fechas_ordenadas)

# Crear el gráfico de barras con color personalizado
plt.figure(figsize=(15, 8))
bars = plt.bar(fechas, conteos, color='skyblue')  # Cambiar el color de las barras a azul claro

# Añadir etiquetas y título
plt.xlabel('Fecha')
plt.ylabel('Número de Artículos')
plt.title('Número de Artículos por Fecha')
plt.xticks(rotation=90)  # Rotar las etiquetas de las fechas para mejor legibilidad

# Mostrar los valores de las barras en el gráfico
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

plt.tight_layout()
plt.show()

```
